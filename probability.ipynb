A priori probability *P(E)*

a random event *E* is defined as the ratio between the number of *s* of the favoralbe cases and the number *n* of the possible cases, which are all considered equally probable:

$P(E) = \frac{s}{n}$

Complementary events

Two events that are mutaully exclusive *E* or $\bar{e}$

Relative frequency and probability

The relative frequency $f (E)$ of an event subjected to $n$ experiments, all carried out under the same conditions, is he ratio between the number $v$ of the times event occurred and the number $n$ of tests carried out:

$f(E) = \frac{v}{n}$

```python
## Uniform distribution

import numpy as np
import matplotlib.pyplot as plt
a = 1
b = 100
N = 100

X1 = np.random.uniform(a,b,N)
plt.plot(X1)
plt.show()
```

```python
plt.figure()
plt.hist(X1, density=True, histtype='stepfilled', alpha=0.2)
plt.show()
```

## Binomial Distribution

$P_{x} = (\frac{n}{x}) p^x q^{n-x}, 0 \leq X \leq n $

* $P_{x}$ is the probability density
* $n$ is the number of independent experiments
* $x$ is the number of successes
* $p$ is the probability of successes
* $q$ is the probability of fail

```python
N = 1000
n = 10
p = 0.5
P1 = np.random.binomial(n, p, N)
plt.plot(P1)
plt.show()
```

```python
plt.figure()
plt.hist(P1, density=True, histtype='bar', alpha=0.8, color = 'green', ec = 'black')
plt.show()
```

## Normal Distribution

$f(X) = \frac{1}{\sqrt{2\pi\sigma}}^{e^{(1/2)[(X-u)/\sigma]^2}}$

* $u$ is the expected value
* $\sigma$ is the standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
mu = 10
sigma = 2
p1 = np.random.normal(mu, sigma, 1000)

mu = 5
sigma = 2
p2 = np.random.normal(mu, sigma, 1000)

mu = 15
sigma = 2
p3 = np.random.normal(mu, sigma, 1000)

mu = 10
sigma = 2
p4 = np.random.normal(mu, sigma, 1000)

mu = 10
sigma = 1
p5 = np.random.normal(mu, sigma, 1000)

mu = 10
sigma = 0.5
p6 = np.random.normal(mu, sigma, 1000)


plot1 = sns.distplot(p1)
plot2 = sns.distplot(p2)
plot3 = sns.distplot(p3)
plot4 = sns.distplot(p4)
plot5 = sns.distplot(p5)
plot6 = sns.distplot(p6)
plt.show()
```
